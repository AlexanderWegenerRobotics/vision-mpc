```python
import sys, os

notebook_dir = os.getcwd()
project_root = os.path.abspath(os.path.join(notebook_dir, '..'))
sys.path.insert(0, project_root)

import numpy as np
from matplotlib import pyplot as plt
from src.common.data_logger import DataLogger
```

```python
path = os.path.join(project_root, 'log/debugging/')
data = DataLogger.load_all(path)

print(data.keys())
print(data["ctrl_arm"].keys())
```

```python
arm = data["arm"]
arm_time = data["arm"]["timestamps"]
arm_time = arm_time - arm_time[0]

ctrl = data["ctrl_arm"]
ctrl_time = data["arm"]["timestamps"]
ctrl_time = ctrl_time - ctrl_time[0]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))

# First subplot - positions
ax1.plot(ctrl["q"])
ax1.plot(ctrl["q_target"])
ax1.set_ylabel('Position (q)')
ax1.grid()
ax1.legend(['q actual', 'q target'])

# Second subplot - torques
ax2.plot(ctrl["tau"])
ax2.set_ylabel('Torque Command')
ax2.set_xlabel('Sample')
ax2.grid()

plt.tight_layout()
plt.show()
```
