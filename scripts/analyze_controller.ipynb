```python
import sys, os

notebook_dir = os.getcwd()
project_root = os.path.abspath(os.path.join(notebook_dir, '..'))
sys.path.insert(0, project_root)

import numpy as np
from matplotlib import pyplot as plt
from src.common.data_logger import DataLogger
```

```python
path = os.path.join(project_root, 'log/test_controller/')
bundle = 'control'
data = DataLogger.load(path, bundle)

print(data.keys())

time = data["timestamps"] - data["timestamps"][0]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))

# First subplot - positions
ax1.plot(time, data["q"])
ax1.plot(time, data["q_target"])
ax1.set_ylabel('Position (q)')
ax1.grid()
ax1.legend(['q actual', 'q target'])

# Second subplot - torques
ax2.plot(time, data["tau_cmd"])
ax2.set_ylabel('Torque Command')
ax2.set_xlabel('Sample')
ax2.grid()

plt.tight_layout()
plt.show()
```
